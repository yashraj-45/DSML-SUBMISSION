```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
heart = pd.read_csv("/content/Heart.csv")
heart.head
```

SCATTERPLOT

```python
#SCATTERPLOT OF A DATASET
plt.scatter(heart["Age"],heart["RestBP"])
c=['b','g','r','c','y','m','orange','k']
#plt.scatter(day,no,color = c,s=size,alpha= 1)
plt.colorbar()
plt.xlabel("Age")
plt.ylabel("RestBP")
plt.title("Scatterplot",fontsize = 20)
plt.show()
```

BARGRAPH

```python
plt.bar(heart["Age"],heart["MaxHR"],width =0.4, color= 'c', edgecolor ='c', alpha = 1 )
plt.xlabel("Age")
plt.ylabel("MaxHR")
plt.title("BAR GRAPH",fontsize = 20)
#plt.bar(X,Y,width =0.4, color= c, edgecolor ='c', alpha = 1 ,label= l)
plt.legend()
plt.show()


```

BOX PLOT

```python
sns.boxplot(y='Chol',data = heart , width = 0.5)
```

```python
sns.boxplot(x='Sex',y= 'Oldpeak',data = heart,showmeans = True,width = 0.5)
```

HISTOGRAM

```python
plt.hist(heart["MaxHR"],bins=100)
plt.xlabel("values")
plt.ylabel("frequency")
plt.title("MAXHR HISTOGRAM")
plt.show()
```

LINE GRAPH

```python
plt.plot(heart["Chol"],heart["MaxHR"])
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("LINE GRAPH")
plt.show()
```

PIE CHART

```python
import matplotlib.pyplot as plt



x = heart["Chol"]
y = heart["Age"]
c = ['b', 'g', 'r', 'c', 'y']

plt.pie(x, labels=y, colors=c, autopct="%0.1f%%", radius=1.5)
plt.legend()
plt.show()
```

```python
sns.set(style="whitegrid")
sns.kdeplot(heart['Age'], shade=True)
plt.title('Density Plot')
plt.xlabel('X Axis Label')
plt.ylabel('Density')
plt.show()
```

COUNTPLOT

```python
sns.set(style="whitegrid")
sns.countplot(x='MaxHR', data=heart)
plt.title('Count Plot')
plt.xlabel('X Axis Label')
plt.ylabel('Count')
plt.show()
```

HEAT PLOT

```python

sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))  # Adjust the figure size as needed

# Assuming 'data' is a DataFrame
sns.heatmap(heart.corr(),annot = True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap')
plt.show()
```
