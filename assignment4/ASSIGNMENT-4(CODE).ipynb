```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
cluster = []
for k in range (1,11):
  kmean =KMeans(n_clusters=k).fit(df)
  cluster.append(kmean.inertia_)
```

```python
plt.plot(range(1,11), cluster ,marker="o")
plt.xlabel("Clusters")
plt.ylabel("inertia")
plt.show()
```

```python
km = KMeans(n_clusters=5).fit(df)
df['Labels']= km.labels_
df.head()
```

```python
df.head()
```

```python
centers = np.array(km.cluster_centers_)
```

```python
colors = {0:'red',1:'green',2:'orange',3:'yellow',4:'pink'}
plt.figure(figsize=(12,8))
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'],linewidths=0.5, c=df['Labels'].map(colors))
plt.title("KMeans clustering")
plt.show()
```
